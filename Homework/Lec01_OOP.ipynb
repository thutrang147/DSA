# Problem 1: Representing a Circle #

Write a Python program to create a class representing a Circle. Include methods to calculate its area and perimeter.
**Class `Circle`**:
   - **Attributes** (nonpublic):
     - `radius`: Radius of the circle (numberic data type only).

   - **Methods**:
     - Getter methods for each attribute.
     - `calculate_circle_area`: Calculate and return the area of the circle.
     - `calculate_circle_perimeter`: Calculate and return the perimeter (circumference) of the circle.

```python
import math

class Circle:
    def __init__(self, radius):
        if not isinstance(radius, (float, int)):
            raise TypeError("The radius must be in numberic data type only.")
        elif radius <= 0:
            raise ValueError("The radius must be a positive number.")
        else:
            self.radius = radius
    
    # getter method
    def get_radius(self):
        return self.radius
    
    def calculate_circle_area(self):
        return round(math.pi * self.radius ** 2, 2)
    
    def calculate_circle_perimeter(self):
        return round(2 * math.pi * self.radius, 2)
    
```

```python
circle1 = Circle(3)
print(circle1.get_radius())
print(circle1.calculate_circle_area())
print(circle1.calculate_circle_perimeter())
```

# Problem 2: Create simple Statistics Module #

Python has the module called statistics and we can use this module to do all the statistical calculations. However, to learn how to make function and reuse function let us try to develop a program, which calculates the measure of central tendency of a sample (mean, median, mode) and measure of variability (range, variance, standard deviation). In addition to those measures, find the min, max, count, percentile, and frequency distribution of the sample. You can create a class called Statistics and create all the functions that do statistical calculations as methods for the Statistics class. Check the output below.

```python
class Statistics:
    def __init__(self, data):
        self.data = data
    
    def count(self):
        return len(self.data)
    
    def sum(self):
        return sum(self.data)
    
    def min(self):
        return min(self.data)
    
    def max(self):
        return max(self.data)
    
    def range(self):
        return (max(self.data) - min(self.data))
    
    def mean(self):
        return sum(self.data) / len(self.data)
    
    def median(self):
        new_data = sorted(self.data)
        n = len(self.data)
        if n % 2 != 0:
            return new_data[n // 2]
        else:
            return (new_data[(n // 2 - 1)] + new_data[(n // 2)]) / 2
    
    def mode(self):
        most_common = max(set(self.data), key=self.data.count)
        return {'mode': most_common, 'count': self.data.count(most_common)}
    
    def std(self):
        mean = self.mean()
        squared_diff_sum = sum((x - mean) ** 2 for x in self.data)
        variance = squared_diff_sum / len(self.data)
        return round(variance ** 0.5, 1)
    
    def var(self):
        mean = self.mean()
        squared_diff_sum = sum((x - mean) ** 2 for x in self.data)
        return round(squared_diff_sum / len(self.data), 1)
    
    def freq_dist(self):
        freq = {}
        for item in self.data:
            freq[item] = freq.get(item, 0) + 1
        total = len(self.data)
        dist = [(freq[item] / total * 100, item) for item in set(self.data)]
        return sorted(dist, reverse=True)
```

```python
ages = [31, 26, 34, 37, 27, 26, 32, 32, 26, 27, 27, 24, 32, 33, 27, 25, 26, 38, 37, 31, 34, 24, 33, 29, 26]

data = Statistics(ages)

print('Count:', data.count()) # 25
print('Sum: ', data.sum()) # 744
print('Min: ', data.min()) # 24
print('Max: ', data.max()) # 38
print('Range: ', data.range()) # 14
print('Mean: ', data.mean()) # 30
print('Median: ', data.median()) # 29
print('Mode: ', data.mode()) # {'mode': 26, 'count': 5}
print('Standard Deviation: ', data.std()) # 4.2
print('Variance: ', data.var()) # 17.5
print('Frequency Distribution: ', data.freq_dist()) # [(20.0, 26), (16.0, 27), (12.0, 32), (8.0, 37), (8.0, 34), (8.0, 33), (8.0, 31), (8.0, 24), (4.0, 38), (4.0, 29), (4.0, 25)]
```

## 1. Card Class
### Purpose
Represents an individual card with a suit and value.

### Functions to Implement
- **`__init__(self, suit, val)`**  
  Initialize a card with its suit (e.g., `"Hearts"`) and value (e.g., `1` for Ace).

- **`show(self)`**  
  Display the card in the format `"Value of Suit"` (e.g., `"Ace of Spades"`).  
  - Use conditions to convert values:
    - `1 -> "Ace"`
    - `11 -> "Jack"`
    - `12 -> "Queen"`
    - `13 -> "King"`
    - Other values remain as-is.


```python
class Card(object):
    def __init__(self, suit, val):
        self.suit = suit
        self.value = val

    def __repr__(self):
        return self.show()

    def show(self):
        card_values = {
            1:  'Ace',
            11: 'Jack',
            12: 'Queen',
            13: 'King'
        }
        if self.value in card_values:
            val = card_values[self.value]
        else:
            val = self.value
        
        return f"{val} of {self.suit}"
```

```python
card1 = Card("Hearts", 1)
print(card1.show())
```

## 2. Deck Class
### Purpose
Represents a collection of 52 cards.

### Functions to Implement
- **`__init__(self)`**  
  - Initialize the deck by building it with 52 cards.

- **`build(self)`**  
  - Generate all 52 cards by combining 4 suits (`Hearts`, `Clubs`, `Diamonds`, `Spades`) with values `1` to `13`
  - Suggestion: self.cards will contain **Card** objects, e.g self.cards = [Card(Hearts, 1), Card(Spade, 4), ...]

- **`shuffle(self, num=1)`**  
  - You can use any method to ensure that drawing a card is completely random

- **`deal(self)`**  
  - Remove and return the top card from the deck.

```python
import random

class Deck(object):
    def __init__(self):
        self.cards = []
        self.build()

    def show(self):
        for card in self.cards:
            print(card.show())

    def build(self):
        suits = ["Hearts", "Clubs", "Diamonds", "Spades"]
        for suit in suits:
            for value in range(1, 14):
                self.cards.append(Card(suit, value))

    def shuffle(self, num=1):
        for _ in range(num):
            random.shuffle(self.cards)

    def deal(self):
        if len(self.cards) > 0:
            return self.cards.pop()
        else:
            return None
```
