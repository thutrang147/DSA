## **Problem 1: Min Binary Heap (4)**

Implement `MinHeap` class which follows the Binary Min Heap data structure. You can use `Item` class given below:

````python
class Item:
    "Item in Heap."

    def __init__(self, key, value):
        self.key = key
        self.val = value

    def __lt__(self, other):
        return self.key < other.key

    def __gt__(self, other):
        return self.key > other.key

    def __eq__(self, other):
        return self.key == other.key

    def __str__(self):
        return str((self.key, self.val))
````

Your task is complete the following methods in the class `MinHeap` below:



````python
class MinHeap:
    def __init__(self):
        " Initialize an empty MinHeap."
        self._data = [] 

    def __len__(self):
        "Returns the number of items in the heap."
        return len(self._data)

    def is_empty(self):
        "Checks if the heap is empty. "
        return len(self._data) == 0

    def _parent(self, k: int):
        "Returns the index of the parent of the item at index k."
        return (k - 1) // 2

    def _left(self, k):
        "Returns the index of the left child of the item at index k."
        return k * 2 + 1

    def _right(self, k):
        "Returns the index of the right child of the item at index k."
        return k * 2 + 2

    def _has_left(self, k):
        "Checks if the item at index k has a left child."
        if self._data[self._left(k)] is None:
            return False
        return True

    def _has_right(self, k):
        "Checks if the item at index k has a right child."
        if self._data[self._right(k)] is None:
            return False
        return True

    def add(self, item: Item):
        """
        Adds a new item into the heap.
        Note: If the new item’s key has already existed in the heap, raise appropriate exception.
        """
        for existing_item in self._data:
            if existing_item.key == item.key:
                raise ValueError(f"Key {item.key} already exists in the heap.")

        self._data.append(item)

        index = len(self._data) - 1
        while index > 0:
            parent = self._parent(index)
            if self._data[index] < self._data[parent]:
                self._data[index], self._data[parent] = self._data[parent], self._data[index] 
                index = parent
            else:
                break

    def get_min(self):
        "Returns the item with minimum key in the heap."
        if self.is_empty():
            raise ValueError("Heap is empty.")
        return self._data[0]

    def remove_min(self):
        "Removes the item with minimum key in the heap."
        if self.is_empty():
            raise ValueError("Heap is empty.")

        self._data[0], self._data[-1] = self._data[-1], self._data[0]
        min_item = self._data.pop()

        index = 0
        while True:
            left = self._left(index)
            right = self._right(index)
            smallest = index

            if left < len(self._data) and self._data[left] < self._data[smallest]:
                smallest = left
            if right < len(self._data) and self._data[right] < self._data[smallest]:
                smallest = right
            
            if smallest == index:
                break

            self._data[index], self._data[smallest] = self._data[smallest], self._data[index]
            index = smallest
        
        return min_item

    def max_heap_sort(self):
        "Sorts the current min heap to a max heap."
        n = len(self._data)
        for index in range((n - 2) // 2, - 1, -1):
            i = index
            while True:
                left = self._left(i)
                right = self._right(i)
                largest = i

                if left < n and self._data[largest] < self._data[left]:
                    largest = left
                if right < n and self._data[largest] < self._data[right]:
                    largest = right
                
                if largest == i:
                    break

                self._data[i], self._data[largest] = self._data[largest], self._data[i]
                i = largest
                
    def __str__(self):
        "Returns a string representation of the heap."
        ans = []
        for item in self._data:
            ans.append(str(item))
        return "[" + ", ".join(ans) + "]"
    
    def __repr__(self):
        return self.__str__()
````

## **Problem 2: Median of A Numeric Array (3)**
The median is the middle value in an ordered numeric 1D array. If the size of the array is even, there is no middle value and the median is the mean of two middle values.

Given a inordered array, implement a solution which utilize heap data structure to find its median. Note that you cannot use Python built-in heap queue object, try to implement some simple binary heaps to support your solution instead.

````python
def find_median(nums):
    min_heap = MinHeap()
    max_heap = MinHeap()

    for num in nums:
        if not max_heap._data:
            max_heap.add(Item(-num, None))
            continue
        
        if num < -max_heap.get_min().key:
            max_heap.add(Item(-num, None))
        else:
            min_heap.add(Item(num, None))
        
        if len(max_heap._data) > len(min_heap._data) + 1:
            val = -max_heap.remove_min().key
            min_heap.add(Item(val, None))
        elif len(min_heap._data) > len(max_heap._data):
            val = min_heap.remove_min().key
            max_heap.add(Item(-val, None))
    
    if len(max_heap._data) > len(min_heap._data):
        return -max_heap.get_min().key
    else:
        return (-max_heap.get_min().key + min_heap.get_min().key) / 2
````

## **Problem 3: Merge Overlapping Intervals (3)**
Given an array of time intervals where $arr[i] = [start_i, end_i]$, the task is to merge all the overlapping intervals into one and output the result which should have only mutually exclusive intervals.

Examples:
```
Input: arr[] = [[1, 3], [2, 4], [6, 8], [9, 10]]
Output: [[1, 4], [6, 8], [9, 10]]
Explanation: In the given intervals, we have only two overlapping intervals [1, 3] and [2, 4].
             Therefore, we will merge these two and return [[1, 4]], [6, 8], [9, 10]].

Input: arr[] = [[7, 8], [1, 5], [2, 4], [4, 6]]
Output: [[1, 6], [7, 8]]
Explanation: We will merge the overlapping intervals [[1, 5], [2, 4], [4, 6]] into a single interval [1, 6].
```

````python
def merge_intervals(arr):
    arr.sort(key=lambda x: x[0])
    merged = []
    for item in arr:
        if not merged or merged[-1][1] < item[0]:
            merged.append(item)
        else:
            merged[-1][-1] = max(merged[-1][1], item[1])
    
    return merged
````

````python
assert merge_intervals([[1, 3], [2, 4], [6, 8], [9, 10]]) == [[1, 4], [6, 8], [9, 10]]
assert merge_intervals([[7, 8], [1, 5], [2, 4], [4, 6]]) == [[1, 6], [7, 8]]
assert merge_intervals([[1, 4], [4, 5]]) == [[1, 5]]
````

# Problem 4: Implement the Hashing class using the Separate Chaining method (4)

````python
class Node:
    def __init__(self, key, value, next=None):
        self.key = key
        self.value = value
        self.next = next

    def __repr__(self):
        return f"({self.key}, '{self.value}')"

class SepChainHash:
    def __init__(self, size):
        self.size = size
        self.data = [None] * self.size

    def hash(self, key):
        return key % self.size

    def add(self, key, value):
        index = self.hash(key)
        current = self.data[index]
        
        prev = None
        while current:
            if current.key == key:
                current.value = value
                return
            prev = current
            current = current.next
        
        new_node = Node(key, value)
        if prev:
            prev.next = new_node
        else:
            self.data[index] = new_node

    def __getitem__(self, key):
        index = self.hash(key)
        current = self.data[index]

        while current:
            if current.key == key:
                return current.value
            current = current.next

        return None

    def delete(self, key):
        index = self.hash(key)
        current = self.data[index]
        previous = None

        while current:
            if current.key == key:
                if previous:
                    previous.next = current.next
                else:
                    self.data[index] = current.next
                return
            previous = current
            current = current.next

    def __repr__(self):
        result = []
        for chain in self.data:
            current = chain
            while current:
                result.append(str(current))
                current = current.next
        return "[" + ", ".join(result) + "]"
````

# Problem 5: Implement a Hashing class with Linear Probing method (4)

````python
class LinProbHash:
    def __init__(self, size):
        self.size = size
        self.data = [None] * self.size

    def hash(self, key):
        return key % self.size

    def add(self, key, value):
        index = self.hash(key)
        original_index = index
        
        while self.data[index] is not None and self.data[index][0] != key:
            index = (index + 1) % self.size
           
            if index == original_index:
                raise ValueError("Hash table is full")
                
        self.data[index] = (key, value)

    def __getitem__(self, key):
        index = self.hash(key)
        original_index = index
        
        while self.data[index] is not None:
            if self.data[index][0] == key:
                return self.data[index][1]
            index = (index + 1) % self.size
            if index == original_index:
                break
                
        return None

    def delete(self, key):
        index = self.hash(key)
        original_index = index
        
        while self.data[index] is not None:
            if self.data[index][0] == key:
                self.data[index] = None
                
                next_index = (index + 1) % self.size
                while self.data[next_index] is not None:
                    key_to_rehash, value_to_rehash = self.data[next_index]
                    self.data[next_index] = None
                    self.add(key_to_rehash, value_to_rehash)
                    next_index = (next_index + 1) % self.size
                    if next_index == original_index:
                        break
                return
                
            index = (index + 1) % self.size
            if index == original_index:
                break

    def __repr__(self):
        return str(self.data)
````

# Problem 6: Mystery of the Cryptic Manuscript (2)

You are an explorer who has discovered a mysterious manuscript filled with sequences of letters. This manuscript contains secrets that can only be revealed by understanding the longest sequence of its cryptic text that uses no more than two distinct letters. Your task is to decode the longest such sequence to help unravel the manuscript's mysteries.

Given a string `s` representing a segment of the manuscript, return the length of the longest substring that contains at most two distinct characters. The success of your mission depends on this crucial piece of information!

### Constraints:
- `s` consists of English letters (lowercase and uppercase).


````python
def longest_substring_two_distinct(s: str) -> int:
    if not s:
        return 0
    
    char_count = {}
    max_length = start = 0
    
    for end, char in enumerate(s):
        char_count[char] = char_count.get(char, 0) + 1
        
        while len(char_count) > 2:
            start_char = s[start]
            char_count[start_char] -= 1
            if char_count[start_char] == 0:
                del char_count[start_char]
            start += 1
        
        max_length = max(max_length, end - start + 1)
    
    return max_length
````
