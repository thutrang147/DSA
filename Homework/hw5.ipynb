## Part 1: Stack Implementation (2)
Implement class `Stack` you learned in the class again with below structure:
- `__init__(self)`: which initialize an empty stack
- `__len__(self)`: return the number of elements in the stack
- `__str__(self)`: Return string presentation of the stack
- `is_empty(self)`: Return True if the stack is empty and vice versa
- `push(self, a)`: Add element a to the top of the stack
- `top(self)`: Return (but do not remove) the element at the top of the stack
- `pop(self)`: Remove and return the element from the top of the stack

````python
class Stack:
    def __init__(self):
        self.data = []
    
    def __len__(self):
        return len(self.data)
    
    def __str__(self):
        return str(self.data)
    
    def is_empty(self):
        return len(self.data) == 0
    
    def push(self, a):
        self.data.append(a)

    def top(self):
        return self.data[-1]
    
    def pop(self):
        return self.data.pop()
    
    
````

````python
A = Stack()
A.push(4)
A.push(0)
A.push(1)
A.push(5)
A.push(9)
A.push(8)
A.push(2)
````

````python
assert A.is_empty() == False
assert str(A) == '[4, 0, 1, 5, 9, 8, 2]'
assert len(A) == 7
assert A.top() == 2
assert A.pop() == 2
````

## Part 2: Moving Elements (1)
Build a function to move all elements from stack A to stack B without changing the order

````python
def moving_elements(stackA: Stack, stackB: Stack) -> Stack:
    temp = Stack()

    while not stackA.is_empty():
        temp.push(stackA.pop())

    while not temp.is_empty():
        stackB.push(temp.pop())

    return stackB
````

# Problem 3: Evaluation of Postfix Expression (2)
Given a postfix expression, the task is to evaluate the postfix expression. A Postfix expression is of the form “a b operator” (“a b +”) i.e., a pair of operands is followed by an operator.

**Examples:**
```
Input: arr = [“2”, “3”, “1”, “*”, “+”, “9”, “-“]
Output: -4
Explanation: If the expression is converted into an infix expression, it will be 2 + (3 * 1) – 9 = 5 – 9 = -4.
```
```
Input: arr = [“100”, “200”, “+”, “2”, “/”, “5”, “*”, “7”, “+”]
Output: 757
Explanation: If the expression is converted into an infix expression, it will be ((100 + 200) / 2) * 5 + 7  = 150 * 5 + 7 = 757.
```
Complete the function below (**Note**: Handle all cases, raise type error if postfix expression is invalid)

````python
def evaluate_postfix(arr):
    if not arr:
        return TypeError("Invalid")
    stack = Stack()
    operators = ['+', '-', '*', '/']

    for element in arr:
        if element.isdigit():
            stack.push(int(element))
        else:
            if element in operators:
                val2 = stack.pop()
                val1 = stack.pop()

                if element == '-':
                    stack.push(val1 - val2)
                elif element == '+':
                    stack.push(val1 + val2)
                elif element == '*':
                    stack.push(val1 * val2)
                elif element == '/':
                    stack.push(val1 / val2)
            else:
                return TypeError("Invalid")
            
    return stack.pop()
````

````python
assert evaluate_postfix(["2", "3", "1", "*", "+", "9", "-"]) == -4
assert evaluate_postfix(["100", "200", "+", "2", "/", "5", "*", "7", "+"]) == 757
````

# Problem 4: Valid Parentheses(1)

Given a string s containing just the characters '(', ')', '{', '}', '[' and ']', determine if the input string is valid.

An input string is valid if:

1. Open brackets must be closed by the same type of brackets.
2. Open brackets must be closed in the correct order.
3. Every close bracket has a corresponding open bracket of the same type.

Example 1:
```
Input: s = "()[]{}"
Output: True

```
Example 2:
```
Input: s = "(]"
Output: False
```

````python
def isValid(s):
    """
    :type s: str
    :rtype: bool
    """
    pairs = {')': '(', '}': '{', ']': '['}
    stack = Stack()
    
    for char in s:
        if char in '[({':
            stack.push(char)
        elif char in '])}':
            if stack.is_empty():
                return False
            if stack.pop() != pairs[char]:
                return False
    
    return stack.is_empty()
        
    
````

````python
assert isValid("()[]{}") == True
assert isValid('(]') == False
````
