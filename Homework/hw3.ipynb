# Problem 2: Counting Digits (1)

Write a recursive function `count_digits(n)` that takes a positive integer n and returns the number of digits in n.

You are not allowed to use loops, string conversions, or in-built functions to calculate the digit count.


````python
def count_digits(n):
    if n < 10:
        return 1
    return 1 + count_digits(n // 10)
````

````python
assert count_digits(12345) == 5
assert count_digits(9) == 1
````

# Problem 3: Palindrome Check (2)

A palindrome is a word, phrase, number, or other sequence of characters that reads the same forward and backward.

Example:
- **Radar** is a palindrome
- **hello** is not a palindrome

Write a recursive function to determine a palindrome.

Hint: To determine a palindrome using recursion, the outer characters of a word are compared. If they match, the inner substring is checked recursively.


````python
def is_palindrome(word) -> bool:
   if len(word) <= 1:
      return True
   if word[0].lower() != word[-1].lower():
      return False
   return is_palindrome(word[1:-1])
````

````python
assert is_palindrome("Radar") == True
assert is_palindrome("hello") == False
assert is_palindrome("123 isi 321") == True
````

# Problem 4: Catalan Numbers (2)

The Catalan numbers form a sequence of natural numbers that occur in various counting problems, often involving recursive structures. The recursive formula for the $n$-th Catalan number is

$$C_n = \sum_{i=0}^{n-1} C_i ⋅ C_{n-1-i}$$

where the base case is $C_0 = 1$

Write a recursive function to determine the $n$-th Catalan number.

````python
def catalan_number_memoized(n, memo={}):
    if n == 0:
        return 1
    return sum(catalan_number_memoized(i) * catalan_number_memoized(n - 1 - i) for i in range(n))
````

````python
assert catalan_number_memoized(0) == 1, "Test case failed: C(0) should be 1"
````

# Problem 5: Generating Valid Parentheses (1)

Write a recursive function `generate_parentheses(n)` that takes an integer n and returns a list of all valid combinations of n pairs of parentheses.

You must ensure that only valid parentheses combinations are generated (e.g., no "())(" or "(())())").

Example:
```
Input: n = 2
Output: ['(())', '()()']
Note:['()()', '(())'] also a valid answer, do not worry about the order.
```

````python
def generate_parentheses(n):
    stack = []
    res = []

    def backtrack(open, closed):
        if open == closed == n:
            res.append("".join(stack))
            return
        
        if open < n:
            stack.append("(")
            backtrack(open + 1, closed)
            stack.pop()
        
        if closed < open:
            stack.append(")")
            backtrack(open, closed + 1)
            stack.pop()


    backtrack(0, 0)
    return res

   
````

````python
assert sorted(generate_parentheses(2)) == ['(())', '()()']
assert sorted(generate_parentheses(3)) == ['((()))', '(()())', '(())()', '()(())', '()()()']
````
