2. List the features of OOP and explain their idea
3. Distinguish between Class and Instance.

**ANSWER HERE:**

1. A programming pattern that rounds around an object or entity are called object-oriented programming.

2. 
- Class: A class is a blueprint for creating objects. Defining attributes (data) and methods (functions) that the objects created from the class will have.

3.

- Polymorphism: Enabling objects to be treated as instances of their parent class, while executing the child class’s implementation. Methods with the same name can perform different actions depending on the object or context.

4.

- Instance is a concrete realization of a class. It is an actual object created using the class and has its own unique data while adhering to the class structure.


# Problem 2: Design Supermarket [4 points]

Implement an OOP system for a supermarket that manages customers, products, orders, and revenue calculations.

**Classes and Specifications:**

1. **Class `Customer`**:
   - **Attributes**:
     - `id` (unique identifier)
     - `name` (customer's name)
     - `dob` (date of birth)
   - **Methods**:
     - `__str__`: Returns a string representation of the customer (e.g., "Customer: John Doe, ID: 123, DOB: 1990-01-01").

2. **Class `Product`**:
   - **Attributes**:
     - `id` (unique identifier)
     - `name` (product name)
     - `description` (brief description of the product)
     - `price` (product price)
   - **Methods**:
     - `__str__`: Returns a string representation of the product.
     - `update`: Updates product details (e.g., price or description).

3. **Class `Order`**:
   - **Attributes**:
     - `customer` (Customer object representing the customer who placed the order)
     - `date_time` (timestamp of when the order was placed)
     - `amount` (total amount of the order)
   - **Methods**:
     - `add_product`: Adds a product to the order.
     - `__str__`: Returns a string representation of the order.

4. **Class `Supermarket`**:
   - **Attribute**:
     - `order_list` (a list of Order objects)
   - **Methods**:
     - `calculate_revenue(month, year)`: Calculates total revenue for the given month and year.
     - `top_selling_product(month, year, k)`: Identifies the top `k` selling products in a given month and year.
     - `search_customer_birthday(month, year)`: Returns a list of customers whose birthdays fall in the specified month and year.

```python
class Customer:
    def __init__(self, id, name, dob):
        self.id = id
        self.name = name
        self.dob = datetime.strptime(dob, "%d-%m-%Y")
    
    def __str__(self):
        return f"Customer: {self.name}, ID: {self.id}, DOB: {self.dob.strftime('%Y-%m-%d')}"
```

```python
class Product:
    def __init__(self, id, name, description, price):
        self.id = id
        self.name = name
        self.description = description
        self.price = price

    def __str__(self):
       return f"Product: {self.name}, ID: {self.id}, Description: {self.description}, Price: {self.price}"
    
    def update(self, name=None, description=None, price=None):
        if name is not None:
            self.name = name
        if description is not None:
            self.description = description
        if price is not None:
            self.price = price
    
```

```python
class Order:
    def __init__(self, customer, date_time):
        self.customer = customer
        self.date_time = datetime.strptime(date_time, "%d-%m-%Y %H:%M:%S")
        self.amount = 0
        self.products = []

    def add_product(self, product, quantity=1):
        self.products.append((product, quantity))
        self.amount += product.price * quantity

    def __str__(self):
        return f"Order by {self.customer.name} on {self.date_time.strftime('%Y-%m-%d %H:%M:%S')}, Total: ${self.amount:.2f}"
```

```python
class Supermarket:
    def __init__(self):
        self.order_list = []
        
    def add_order(self, order):
        self.order_list.append(order)
        
    def calculate_revenue(self, month, year):
        return sum(order.amount for order in self.order_list 
                  if order.date_time.month == month and order.date_time.year == year)
    
    def top_selling_product(self, month, year, k):
        product_sales = defaultdict(int)
        for order in self.order_list:
            if order.date_time.month == month and order.date_time.year == year:
                for product, quantity in order.products:
                    product_sales[product.name] += quantity
        return sorted(product_sales.keys(), key=lambda x: product_sales[x], reverse=True)[:k]
    
    def search_customer_birthday(self, month=None, year=None):
        customers = []
        for order in self.order_list:
            if ((month is None or order.customer.dob.month == month) and 
                (year is None or order.customer.dob.year == year) and 
                order.customer not in customers):
                customers.append(order.customer)
        return customers
```

```python
# Create Customer
customer1 = Customer(1, "Alice", "10-04-1994")
customer2 = Customer(2, "Bob", "15-06-1994")

# Creating Products
product1 = Product(101, "Milk", "1 Litre Organic Milk", 2.99)
product2 = Product(102, "Bread", "Whole Wheat Bread", 3.50)
product3 = Product(103, "Apples", "Fresh Apples", 4.00)

# Creating Orders
order1 = Order(customer1, "10-01-2024 10:00:00")
order1.add_product(product1, 2)
order1.add_product(product2, 3)

order2 = Order(customer2, "15-01-2024 10:00:00")
order2.add_product(product2)
order2.add_product(product3)

# Updating Product
product2.update(name="Butter", description="Salted Butter", price=5.00)

# Print Order Details
print(order1)
print(order2)

# Creating Supermarket and Adding Orders
supermarket = Supermarket()
supermarket.add_order(order1)
supermarket.add_order(order2)

# Displaying Supermarket Statistics
print("Total Revenue in January 2024:", supermarket.calculate_revenue(1, 2024))
print("Top Selling Product in January 2024:", supermarket.top_selling_product(1, 2024, 2))

assert order1.amount == 16.48
assert order2.amount == 7.5
assert supermarket.calculate_revenue(1, 2024) == 23.98
assert supermarket.top_selling_product(1, 2024, 2) == ['Butter', 'Milk']
assert supermarket.top_selling_product(1, 2024, 1) == ['Butter']
```

```python
# Test case search customer birthday
customer_list = supermarket.search_customer_birthday(4, 1994)
assert len(customer_list) == 1
assert [customer.name for customer in customer_list] == ['Alice']

customer_list = supermarket.search_customer_birthday(month=6)
print([customer.name for customer in customer_list])
assert len(customer_list) == 1
assert [customer.name for customer in customer_list] == ['Bob']

customer_list = supermarket.search_customer_birthday(year=1994)
print([customer.name for customer in customer_list])
assert len(customer_list) == 2
assert [customer.name for customer in customer_list] == ['Alice', 'Bob']

customer_list = supermarket.search_customer_birthday(1, 1995)
assert len(customer_list) == 0
assert [customer.name for customer in customer_list] == []
```

# Problem 3: Design Browser History [2 points]

You have a **browser** of one tab where you start on the `homepage` and you can visit another `url`, get back in the history number of `steps` or move forward in the history number of `steps`.

**Implement the `BrowserHistory` class:**

1. **Constructor - `__init__(self, homepage)`**:
   - Initializes the browser with the given homepage.
   - Parameters:
     - `homepage` (string): The homepage URL of the browser.

2. **Method - `visit(self, url)`**:
   - Visits the given URL from the current page and updates the history.
   - Clears all forward history (i.e., if you visit a new page after going back in history, the forward history is erased).
   - Parameters:
     - `url` (string): The URL to visit.

3. **Method - `back(self, steps)`**:
   - Moves back in the history by a specified number of steps.
   - If the number of steps is more than the available history, move back as far as possible.
   - Returns the current URL after moving back.
   - Parameters:
     - `steps` (int): The number of steps to go back in history.

4. **Method - `forward(self, steps)`**:
   - Moves forward in the history by a specified number of steps.
   - If the number of steps is more than the available forward history, move forward as far as possible.
   - Returns the current URL after moving forward.
   - Parameters:
     - `steps` (int): The number of steps to go forward in history.

```python
class BrowserHistory:
    def __init__(self, homepage: str):
        self.homepage = homepage
        self.history = [homepage]
        self.current = 0

    def visit(self, url: str):
        self.current += 1
        self.history = self.history[:self.current]
        self.history.append(url)
    
    def back(self, steps: int):
        if self.current - steps >= 0:
            self.current -= steps
        else:
            self.current = 0
        
        return self.history[self.current]
    
    def forward(self, steps: int):
        if self.current + steps <= len(self.history) - 1:
            self.current += steps
        else:
            self.current = len(self.history) - 1
        
        return self.history[self.current]
            
```

```python
obj = BrowserHistory("leetcode.com")
assert obj.visit("google.com") == None      # You are in "leetcode.com". Visit "google.com"
assert obj.visit("facebook.com") == None    # You are in "google.com". Visit "facebook.com"
assert obj.visit("youtube.com") == None     # You are in "facebook.com". Visit "youtube.com"
assert obj.back(1) == "facebook.com"        # You are in "youtube.com", move back to "facebook.com" return "facebook.com"
assert obj.back(1) == "google.com"          # You are in "facebook.com", move back to "google.com" return "google.com"
assert obj.forward(1) == "facebook.com"     # You are in "google.com", move forward to "facebook.com" return "facebook.com"
assert obj.visit("linkedin.com") == None    # You are in "facebook.com". Visit "linkedin.com"
assert obj.forward(2) == "linkedin.com"     # You are in "linkedin.com", you cannot move forward any steps.
assert obj.back(2) == "google.com"          # You are in "linkedin.com", move back two steps to "facebook.com" then to "google.com". return "google.com"
assert obj.back(7) == "leetcode.com"        # You are in "google.com", you can move back only one step to "leetcode.com". return "leetcode.com"
```
