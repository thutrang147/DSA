```python
def binary_search(arr, low, high, x):
    if high >= low:
        mid = low + (high - low) // 2

        if arr[mid] == x:
            return True
        
        elif arr[mid] > x:
            return binary_search(arr, low, mid-1, x)
        else:
            return binary_search(arr, mid + 1, high, x)
    
    else:
        return -1

arr = sorted([9, 8, 10, 45, 63, 7])
print(binary_search(arr, 0, len(arr)-1, 7))
```

```python
def fibonacci(n, mem = {}):
    if n == 0 or n == 1:
        return n
    else:
        if n-1 not in mem:
            mem[n-1] = fibonacci(n-1)
        if n-2 not in mem:
            mem[n-2] = fibonacci(n-2)
        return mem[n-1] + mem[n-2]


print(fibonacci(100))
```
