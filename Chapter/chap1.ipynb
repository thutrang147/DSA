```python
#R-1.1
def is_multiple(n, m):
    if n % m == 0:
        return True
    else: 
        return False
    
print(is_multiple(30, 5))
```

```python
#R-1.2
def is_even(k):
    even = list('02468')
    num = str(k)
    if num[-1] in even:
        return True
    else:
        return False
    
print(is_even(12))
print(is_even(13))
    
```

```python
#R-1.3
def minmax(data):
    if not data:  
        return None
    minn = maxx = data[0]  
    for num in data:
        if num < minn:
            minn = num
        if num > maxx:
            maxx = num
    
    return (minn, maxx) 

print(minmax([1, 3, 4, 5, 6, 11, 15]))
```

```python
#R-1.4
def sum(n):
    ans = 0
    for i in range(1, n):
        ans += i**2

    return ans

print(sum(3))
```

```python
#R-1.6
def odd_sum(n):
    ans = 0
    for i in range(1, n):
        if i % 2 != 0:
            ans += i**2

    return ans

print(odd_sum(5))
```

```python
def reverse_list(input_list):
    return input_list[::-1]

print(reverse_list([1, 2, 3]))
```

```python
def has_odd_product_pair(sequence):
    odd_numbers = [num for num in sequence if num % 2 != 0]
    return len(odd_numbers) >= 2

sequence = [2, 4, 6, 7, 9]
print(has_odd_product_pair(sequence))  
```

```python
def distinct(numbers):
    temp = 1
    for num in numbers:
        temp = temp ^ num
    
    return temp == 0

print(distinct([1, 3, 4, 5, 2, 1, 3]))
```
