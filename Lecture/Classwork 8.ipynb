# Given Binary Tree

```python
class Node:
    def __init__(self, value, parent=None):

        self.value = value
        self.parent = parent
        self.left = None
        self.right = None


class BinaryTree:
    def __init__(self, root = None):
        self.root = root

    def add_left(self, p_node: Node, c_node: Node):
        p_node.left = c_node
        c_node.parent = p_node

    def add_right(self, p_node: Node, c_node: Node):
        p_node.right = c_node
        c_node.parent = p_node

    def is_root(self, node: Node):
        return self.root == node

    def is_leaf(self, node: Node):
        if not node:
            return False
        return node.left == None and node.right == None

    def check_ancestor(self, node1: Node, node2: Node):
        if node1 == node2: return False
        else:
            ptr = node2
            while ptr:
                if node1 == ptr:
                    return True
                ptr = ptr.parent
            return False

    def depth(self, node: Node):
        if node.parent == None:
            return 0
        return 1 + self.depth(node.parent)

    def height(self, node: Node):
        if node == None:
            return 0
        if self.is_leaf(node):
            return 0
        return 1 + max(self.height(node.left), self.height(node.right))

    def sum_right_leaves(self, node: Node):
        if self.is_leaf(node):
            if node.parent == None:
                return 0
            else:
                if node.parent.right == node:
                    return node.value
        if not node: return 0
        else:
            return self.sum_right_leaves(node.left) + self.sum_right_leaves(node.right)

    def __str__(self):
        return self._print_tree(self.root, 0)

    def _print_tree(self, node, depth):
        if node is None:
            return ""

        right_part = self._print_tree(node.right, depth + 1)
        node_part = "   " * depth + str(node.value) + "\n"
        left_part = self._print_tree(node.left, depth + 1)
        return right_part + node_part + left_part

```

# Problem 1: Path to the target

Given the root of a binary tree and an integer targetSum, return all root-to-leaf paths where the sum of the node values in the path equals targetSum. Each path should be returned as a list of the node values, not node references.

A root-to-leaf path is a path starting from the root and ending at any leaf node. A leaf is a node with no children.

# Problem 2: Nearest Ancestor of a Binary Tree
Given a binary tree, find the lowest common ancestor (LCA) of two given nodes in the tree.

According to the definition of LCA on Wikipedia: “The lowest common ancestor is defined between two nodes p and q as the lowest node in T that has both p and q as descendants (where we allow a node to be a descendant of itself).”

```python
def lowestCommonAncestor(root, p, q):
    """
    input:
        root: root of the tree
        p: node p
        q: node q
    
        return:
            node of the ancestor of both p and q
     
    """
    if not root:
        return None
    if root.value == p.value or root.value == q.value:
        return root
    
    lcal = lowestCommonAncestor(root.left, p, q)
    lcar = lowestCommonAncestor(root.right, p, q)

    if lcal and lcar:
        return root
    
    return lcal if lcal else lcar
    
```

```python
# Creating the given tree structure
n1 = Node(1)
n2 = Node(2)
n3 = Node(3)
n4 = Node(4)
n5 = Node(5)
n6 = Node(6)
n7 = Node(7)
n8 = Node(8)
n9 = Node(9)

tree = BinaryTree(n1)
tree.add_left(n1, n2)
tree.add_right(n1, n3)
tree.add_left(n2, n4)
tree.add_right(n2, n5)
tree.add_left(n5, n7)
tree.add_right(n5, n8)
tree.add_left(n3, n6)
tree.add_left(n6, n9)
print(tree)
```

```python
assert lowestCommonAncestor(tree.root, n4, n7) == n2
assert lowestCommonAncestor(tree.root, n4, n9) == n1
assert lowestCommonAncestor(tree.root, n9, n3) == n3
```
