```python
class ArrayQueue:
    DEFAULT_CAPACITY = 0

    def __init__(self):
        self._data = [None] * ArrayQueue.DEFAULT_CAPACITY
        self._size = 0
        self._front = 0
    
    def __len__(self):
        return self._size

    def is_empty(self):
        return self._size == 0

    def first(self):
        if self.is_empty():
            return "Queue is empty"
        return self._data[self._front]
    
    def dequeue(self):
        if self.is_empty():
            return "Queue is empty"
        answer = self._data[self._front]
        self._data[self._front] = None
        self._front = (self._front + 1) % len(self._data)
        self._size -= 1

        return answer

    def enqueue(self, e):
        if self._size == len(self._data):
            self._resize(2 * len(self.data))
        
        avail = (self._front + self._size) % len(self._data)
        self._data[avail] = e
        self._size += 1

    def _resize(self, cap):
        old = self._data
        self._data = [None] * cap
        walk = self._front
        for k in range(self._size):
            self._data[k] = old[walk]
            walk = (1 + walk) % len(old)
        
        self._front = 0
    
        
```

```python
class Queue:
    def __init__(self):
        self._data = []
    
    def is_empty(self):
        return len(self._data) == 0
    
    def enqueue(self, e):
        self._data.append(e)
    
    def dequeue(self):
        if self.is_empty():
            return TypeError("Queue is empty")
        return self._data.pop(0)  

    def len(self):
        return len(self._data)
    
    def first(self):
        if self.is_empty():
            return "error"   
        return self._data[0]  

    def show(self):
        if self.is_empty():
            return "Queue is empty"
        return self._data

```

```python
Q = Queue()
Q.enqueue(5)
Q.enqueue(3)
Q.len()
Q.dequeue()
Q.is_empty()
Q.len()
Q.is_empty()
Q.enqueue(7)
Q.enqueue(9)
Q.first()
Q.enqueue(4)
Q.len()
Q.dequeue()
Q.show()
```

```python
class ArrayStack:
    def __init__(self):
        self._data = []

    def push(self, val):
        self._data.append(val)

    def pop(self):
        if self.is_empty():
            raise Exception('Stack is empty')
        return self._data.pop()

    def top(self):
        if self.is_empty():
            raise Exception('Stack is empty')
        return self._data[-1]

    def is_empty(self):
        return len(self._data) == 0

    def __len__(self):
        return len(self._data)
    
    def show(self):
        return self._data
```

```python
def reverse_queue(queue: Queue):
    stack = ArrayStack()
    
    while not queue.is_empty():
        stack.push(queue.dequeue())
    
    while not stack.is_empty():
        queue.enqueue(stack.pop())
    
    return queue


Q = Queue()
for i in [1, 2, 3, 4, 5]:
    Q.enqueue(i)

reversed_queue = reverse_queue(Q)
print(reversed_queue.show())
        
```

```python
class NewStack:
    def __init__(self):
        self.q1 = Queue()
        self.q2 = Queue()

    def push(self, x):
        self.q2.enqueue(x)
        while(self.q1):
            self.q2.enqueue(self.q1.dequeue())

        self.q1, self.q2 = self.q2, self.q1

    def pop(self):
        if self.q1:
            self.q1.dequeue()

    def top(self):
        if (self.q1):
            return self.q1[0]
        return None

    def size(self):
        return len(self.q1)
    
    def show(self):
        return self.q1
    
```

```python
s = NewStack()
s.push(1)
s.push(2)
s.push(3)

print("current size: ", s.size())
print(s.top())
s.pop()
print(s.top())
s.pop()
print(s.top())

print("current size: ", s.size())
```
