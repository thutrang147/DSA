```python
class ArrayStack:
    def __init__(self):
        self._data = []

    def push(self, val):
        self._data.append(val)

    def pop(self):
        if self.is_empty():
            raise Exception('Stack is empty')
        return self._data.pop()

    def top(self):
        if self.is_empty():
            raise Exception('Stack is empty')
        return self._data[-1]

    def is_empty(self):
        return len(self._data) == 0

    def __len__(self):
        return len(self._data)
    
    def show(self):
        return self._data
```

```python
my_stack = ArrayStack()
my_stack.push(1)
my_stack.show()
```

```python
class WebBrowser():
    def __init__(self, homepage: str):
        self.homepage = homepage
        self.history = [homepage]
        self.current = 0

    def access(self, url: str):
        self.current += 1
        self.history = self.history[:self.current]
        self.history.append(url)

    def back(self):
        if self.current - 1 >= 0:
            self.current -= 1
        else:
            self.current = 0

        return self.history[self.current]

    def next(self):
        if self.current + 1 <= len(self.history) - 1:
            self.current += 1
        else:
            self.current = len(self.history) - 1
        
        return self.history[self.current]
```

```python
class WebBrowser():
    def __init__(self, homepage: str):
        self.homepage = homepage
        self.current = homepage
        self.back_stack = ArrayStack()
        self.forward_stack = ArrayStack()

    def access(self, url):
        self.back_stack.push(self.current)
        self.current = url
        self.forward_stack = ArrayStack()
    
    def back(self):
        if not self.back_stack.is_empty():
            self.forward_stack.push(self.current)
            self.current = self.back_stack.pop()
            return self.current
        
        return self.current
    
    def next(self):
        if not self.forward_stack.is_empty():
            self.back_stack.push(self.current)
            self.current = self.forward_stack.pop()
            return self.current
        
        return self.current
    
    def show(self):
        return self.current
    

            
```

```python
browser = WebBrowser("google.com")
browser.access("youtube.com")
browser.access("github.com")
browser.show()
```

```python
def check_parentheness(n):
    open = ['(', '{', '[']
    close = [')', '}', ']']
    
    save_open = ArrayStack()
    save_close = []
    for letter in n:
        if letter in open:
            save_open.push(letter)
        else:
            save_close.append(letter)
    
    for letter in save_close:
        if letter in close:
            if save_open.is_empty():
                return False
            if close.index(letter) != open.index(save_open.pop()):
                return False
    return save_open.is_empty()



print(check_parentheness("(([])))"))
```
