```python
class CreditCard:
    def __init__(self, customer, bank, acnt, limit):
        self._customer = customer
        self._bank = bank
        self._account = acnt
        self._limit = limit
        self._balance = 0

    def get_customer(self):
        return self._customer
    
    def get_bank(self):
        return self._bank

    def get_account(self):
        return self._account
    
    def get_limit(self):
        return self._limit
    
    def get_balance(self):
        return self._balance
    
    def charger(self, price):
        if price + self._balance > self._limit:
            return False
        else:
            self._balance += price
            return True
        
    def make_payment(self, amount):
        self._balance -= amount


cc = CreditCard( 'John Doe', '1st Bank', '5391 0375 9387 5309', 1000)
```

```python
class Vector:
    def __init__(self, size):
        self._coords = [0] * size

    def __len__(self):
        return len(self._coords)
    
    def __getitem__(self, j):
        return self._coords[j]

    def __setitem__ (self, j, val):
        self._coords[j] = val
    
    def __add__(self, other):
        if len(self) != len(other):
            raise ValueError('dimensions must agree')
        result = Vector(len(self))
        for j in range(len(self)):
            result[j] = self[j] + other[j]
        return result
    
    def __substract__(self, other):
        if len(self) != len(other):
            raise ValueError('dimensions must agree')
        result = Vector(len(self))
        for j in range(len(self)):
            result[j] = self[j] - other[j]
        return result
    
    def __dotproduct__(self, other):
        if len(self) != len(other):
            raise ValueError('dimensions must agree')
        result = 0
        for j in range(len(self)):
            result += self[j] * other[j]
        return result
    
    def __multiply__(self, a):
        result = Vector(len(self))
        for j in range(len(self)):
            result[j] = self[j] * a
        return result
    
    def __repr__(self):
        return str(self._coords)
    
x = Vector(2)
x.__setitem__(0, 1)
x.__setitem__(1, 2)
y = Vector(2)
y.__setitem__(0, 1)
y.__setitem__(1, 2)

print(x)
print(x.__add__(y))
print(x.__substract__(y))
print(x.__dotproduct__(y))
print(x.__multiply__(2))
```
