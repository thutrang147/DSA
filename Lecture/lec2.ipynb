```python
def prefix_average1(S):
    n = len(S)
    A = [0] * n
    for j in range(n):
        total = 0
        for i in range(j+1):
            total += S[i] 
        A[j] = total / (j+1)
    return A

print(prefix_average1([1, 2, 3, 4]))
```

```python
def prefix_average2(S):
    n = len(S)
    A = [0] * n
    for j in range(n):
        A[j] = sum(S[0:j+1]) / (j+1)

    return A

print(prefix_average2([1, 2, 3, 4]))
```

```python
def prefix_average3(S):
    n = len(S)
    A = [0] * n
    total = 0
    for j in range(n):
        total += S[j]
        A[j] = total / (j+1)
    
    return A

print(prefix_average3([1, 2, 3, 4]))
```

```python
def find(numbers, x):
    if x in numbers:
        return True
    return False

print(find([3, 5, 7, 8], 7))
```

```python
def binary_search(numbers, x):
    left = 0
    right = len(numbers) - 1
    while left <= right:
        mid = left + (right - left) // 2

        if numbers[mid] == x:
            return mid
        
        elif numbers[mid] < x:
            left = mid + 1

        else:
            right = mid - 1
        
    return -1


print(binary_search([3, 5, 7, 8], 7))
```
