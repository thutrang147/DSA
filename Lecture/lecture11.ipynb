```python
class DictData:
    def __init__(self, size=10):
        self.data = [None] * size
    
    def hash_key(self, key): # chuyen tat ca key trong ascii r cong lai phan du sang 10 -> tu 0 - 9
        total = 0
        for element in key:
            total += ord(element)
        
        return total % len(self.data)

    def __getitem__(self, key):
        return self.data[self.hash_key(key)]

    def __setitem__(self, key, value):
        self.data[self.hash_key(key)] = value

d = DictData()
d['B'] = 1
d['C'] = 2
d['W'] = 3

print(d['B'])
```
