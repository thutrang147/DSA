Linked List

Chứa 1 list các phần tử không cần thiết phải liên tiếp trên memory.
Mỗi phần tử/ class chứa 2 thành phần: value và next

```python
class Node:
    def __init__(self, element, next = None):
        self.element = element
        self.next = next
        
```

```python
class LinkedList:
    def __init__(self):
        self._head = None
        self._size = 0
    
    def __len__(self):
        return self._size
    
    def is_empty(self):
        return self._size == 0
    
    def add(self, element):
        # self._head = Node(element, self._head)
        if self.is_empty():
            self._head = Node(element)
        else:
            new = Node(element, self._head) 
            self._head = new
        self._size += 1
    
    def __repr__(self):
        start = self._head
        elements = []
        while start:
            elements.append(start.element)
            start = start.next
        
        return '->'.join(map(str, elements))
    
    def add_end(self, element):
        start = self._head
        while start.next:
            start = start.next
        start.next = Node(element)
        self._size += 1
        
        
    def remove_element(self, element):
        start = self._head
        prev = None
        while start:
            if start.element == element:
                if prev:
                    prev.next = start.next
                else:
                    self._head = start.next
                
                self._size -= 1

            prev = start
            start = start.next

```

```python
my_linkedlist = LinkedList()
my_linkedlist.add(3)
my_linkedlist.add(4)
my_linkedlist.add_end(5)
my_linkedlist.remove_element(5)
print(my_linkedlist.__repr__())
```
