### Tree

- Trong cây có khái niệm các Node. Các nốt ở trong cây có quan hệ cha con. 
- Nốt không có cha gọi là nốt gốc.
- Nốt không có con gọi là nốt lá.
- Trong cây không bao giờ có vòng khép kín. 

```python
class Node:
    def __init__(self, element, parent=None, left_child=None, right_child=None):
        self.element = element
        self.parent = parent
        self.left_child = left_child
        self.right_child = right_child

class BinaryTree:
    def __init__(self, root=None):
        self.root = root
    
    def add_left(self, parent, element):
        if parent.left_child is None:
            parent.left_child = Node(element, parent)
        else:
            raise ValueError('Left child exists')
    
    # tất cả con trái nhỏ hơn nốt đấy, tất cả con phải lớn hơn nốt đấy
    
    def search(self, value):
        curr = self.root
        while curr:
            if value == curr.element:  
                return True
            elif value > curr.element:  
                curr = curr.right_child  
            else:
                curr = curr.left_child  
        
        return False  
    

    def add_right(self, parent, element):
        if parent.right_child is None:
            parent.right_child = Node(element, parent)
        else:
            raise ValueError('Right child exists')

    def is_root(self, node):
        return self.root == node
    
    def num_children(self, node):
        count = 0
        if node.left_child:
            count += 1
        if node.right_child:
            count += 1
        return count

    def is_leaf(self, node):
        return self.num_children(node) == 0

    def __str__(self):
        def _str(node):
            if node is None:
                return "None"
            left_str = _str(node.left_child)
            right_str = _str(node.right_child)
            return f"{node.element} -> ({left_str}, {right_str})"
        
        return _str(self.root)
```

```python
root = Node(5)
tree = BinaryTree(root)

tree.add_left(root, 3)
tree.add_right(root, 8)

tree.add_left(root.left_child, 1)
tree.add_right(root.left_child, 4)

tree.add_left(root.right_child, 6)
tree.add_right(root.right_child, 9)

print(str(tree))

print(tree.is_root(root)) 
print(tree.is_leaf(root))  
print(tree.is_leaf(root.left_child.left_child))  
print(tree.is_leaf(root.right_child.right_child)) 

print(tree.search(3))
```
